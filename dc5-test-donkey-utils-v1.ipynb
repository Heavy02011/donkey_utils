```python
import os
import pkg_resources

def list_installed_files(package):
    # Locate the package
    distribution = pkg_resources.get_distribution(package)
    if not distribution:
        return f"No package named {package} found."
    
    # List files associated with the package
    relative_file_paths = distribution.get_metadata_lines('RECORD')
    install_path = distribution.location  # Base installation path

    # Print full paths of the files
    for line in relative_file_paths:
        path = line.split(',')[0]  # Get the file path part from the RECORD entry
        print(os.path.join(install_path, path))

# Use the function to list files
list_installed_files('donkey_utils')
```

```python
import os
import pip

def list_installed_files(package):
    # Locate the package
    package_info = [p for p in pip.get_installed_distributions() if p.project_name.lower() == package.lower()]
    if not package_info:
        return f"No package named {package} found."
    
    # Extract the package installation location
    package_location = package_info[0].location
    
    # Walk through the directory tree, listing contents
    for root, dirs, files in os.walk(package_location):
        for file in files:
            print(os.path.join(root, file))

# Use the function to list files
list_installed_files('donkey_utils')
```
